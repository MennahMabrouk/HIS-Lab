```python
import pydicom
from PIL import Image
import numpy as np
import os
import glob
from matplotlib import cm
from matplotlib import pyplot as plt
```

```python
import matplotlib.pyplot as plt
import pydicom
import pydicom.data

# Full path of the DICOM file is passed in base
base = r"/content/sample_data"
pass_dicom = "exampleDICOM.dcm" # file name is 1-12.dcm

# enter DICOM image name for pattern
# result is a list of 1 element
filename = pydicom.data.data_manager.get_files(base, pass_dicom)[0]

ds = pydicom.dcmread(filename)

plt.imshow(ds.pixel_array, cmap=plt.cm.bone) # set the color map to bone
plt.show()
```

```python
# Load a DICOM file
ds = pydicom.dcmread(r"/content/sample_data/exampleDICOM.dcm")

# Access and print some DICOM attributes
print("Patient Name:", ds.PatientName)
print("Patient ID:", ds.PatientID)
print("Modality:", ds.Modality)
print("Rows:", ds.Rows)
print("Columns:", ds.Columns)
```

```python
# Patient Information
print("Patient Information:")
print("Patient Name:", ds.PatientName)
print("Patient ID:", ds.PatientID)
print("Patient's Birth Date:", ds.PatientBirthDate)
print("Patient's Sex:", ds.PatientSex)
print("Patient's Age:", ds.PatientAge)
```

```python
# Study Information
print("\nStudy Information:")
print("Study Instance UID:", ds.StudyInstanceUID)
print("Study Date:", ds.StudyDate)
print("Study Time:", ds.StudyTime)
print("Accession Number:", ds.AccessionNumber)
```

```python
# Series Information
print("\nSeries Information:")
print("Series Instance UID:", ds.SeriesInstanceUID)
print("Modality:", ds.Modality)
print("Series Description:", ds.SeriesDescription)
```

```python
# Image Information
print("\nImage Information:")
print("SOP Instance UID:", ds.SOPInstanceUID)
print("Rows:", ds.Rows)
print("Columns:", ds.Columns)
print("Pixel Spacing:", ds.PixelSpacing)
print("Bits Allocated:", ds.BitsAllocated)
print("Bits Stored:", ds.BitsStored)
print("Pixel Representation:", ds.PixelRepresentation)
```
